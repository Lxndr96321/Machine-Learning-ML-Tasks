```python
import numpy as np, pandas as pd, seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.columns
```

```python
df = df[['sex', 'age', 'fare','class', 'who', 'embark_town', 'alive', 'alone']]
df
```

```python
df = df.dropna()
```

```python
X, y = df.drop(columns = ['alive']), df['alive']
```

```python
xCat = X.select_dtypes(include = ['object','category', 'bool'])
xCont = X.select_dtypes(exclude = ['object','category', 'bool'])
```

```python
xCatOrdinal = xCat['class']
xCatNominal = xCat.drop(columns = ['class'])
```

```python
# split data to train and test
import sklearn
from sklearn.model_selection import train_test_split
```

```python
xContTrain, xContTest, xCatOrdinalTrain, xCatOrdinalTest, xCatNominalTrain, xCatNominalTest ,yTrain, yTest = train_test_split(xCont, xCatOrdinal, xCatNominal, y,
                                                                             test_size = 0.32,
                                                                             random_state = 23)
```

```python
# cont
# robust scaling --> step -1 (fit_transform on xContTrain & transform on xContTest)
# box - cox ---> step - 2 (fit_transform on xContTrain & transform on xContTest)
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rScaler = RobustScaler()
```

```python
# fit_transform

xContTrain = pd.DataFrame(rScaler.fit_transform(xContTrain),
                          columns = xContTrain.columns,
                          index = xContTrain.index)

xContTrain
```

```python
xContTest = pd.DataFrame(rScaler.transform(xContTest),
                          columns = xContTest.columns,
                          index = xContTest.index)

xContTest
```
