## Weighted KNN Code using different Distances

### Creating a weighted k-NN classifier with different distance metrics in Python involves using the scikit-learn library. A weighted k-NN classifier using three different distance metrics: Euclidean distance, Manhattan distance, and Chebyshev distance. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier

# Generate a synthetic dataset for classification
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, random_state=42)

# Define the number of neighbors (k)
k = 3

# Create a dictionary to store distance metrics and their names
distance_metrics = {
    'Euclidean': 'euclidean',
    'Manhattan': 'manhattan',
    'Chebyshev': 'chebyshev',
}

# Create subplots for visualization
plt.figure(figsize=(15, 5))

# Loop through each distance metric
for i, (metric_name, metric) in enumerate(distance_metrics.items()):
    # Create a weighted KNeighborsClassifier
    weighted_knn = KNeighborsClassifier(n_neighbors=k, weights='distance', metric=metric)
    
    # Fit the model to the data
    weighted_knn.fit(X, y)
    
    # Create a meshgrid for visualization
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    
    # Predict class labels for each point in the mesh
    Z = weighted_knn.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot decision boundaries for the current distance metric
    plt.subplot(1, len(distance_metrics), i + 1)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(f'Weighted KNN ({metric_name} Distance)')
    
plt.tight_layout()
plt.show()
```

- In this code:

- We create a synthetic classification dataset using make_classification from scikit-learn.

- We define the number of neighbors k.

- We create a dictionary distance_metrics that maps distance metric names to their corresponding scikit-learn metric strings.

- We loop through each distance metric, create a weighted KNeighborsClassifier using that metric, fit the model to the data, and plot the decision boundaries for visualization.

- Finally, we use subplots to display the decision boundaries for each distance metric side by side.

- This code will produce a visualization with three subplots, each showing the decision boundaries of the weighted k-NN classifier using a different distance metric (Euclidean, Manhattan, and Chebyshev). You can customize the dataset, the number of neighbors, and other parameters as needed for your specific application.
